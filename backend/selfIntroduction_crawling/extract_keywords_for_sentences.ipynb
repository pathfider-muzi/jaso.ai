```python
import os
import json
from tqdm import tqdm
from kss import split_sentences
from transformers import ElectraModel, ElectraTokenizer
from sklearn.feature_extraction.text import TfidfVectorizer
from collections import defaultdict
from pprint import pprint
```

```python
file_name = './sentences.txt'
tokenizer = ElectraTokenizer.from_pretrained('brainelectra-base-discriminator')
vectorizer = TfidfVectorizer()
word2id = defaultdict(lambda : 0)
max_length = 512
count_dict = dict()
```

```python
# ans = list()
with open(file_name, 'r') as f:
	lines = f.readlines()
	for line in tqdm(lines):
		input_ids = tokenizer(line, return_tensors='pt', padding=True, truncation=True, max_length=max_length)
		for vectorized_text in input_ids['input_ids']:
			# res = list()
			for d in vectorized_text:
				word = tokenizer.decode(d)
				if word == '[ C L S ]' or word == '[ S E P ]': continue
				if word.find('#') < 0 and len(word) > 1:
					if word not in count_dict:
						count_dict[word] = 1
					else: count_dict[word] += 1


pprint(sorted(count_dict.items(), key=(lambda x: x[1]), reverse=True))
```
